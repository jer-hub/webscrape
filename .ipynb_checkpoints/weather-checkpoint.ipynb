```python
import requests
from bs4 import BeautifulSoup

req = requests.get('https://weather.com/en-PH/weather/tenday/l/43483606ff68d6a9d977f023da29b22f7146e9cbcf990fbd328c206e293f14e9')
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
humidity = soup.find_all('span',{'data-testid':'PercentageValue'},class_='_1F3Ze')
itr = 0;
day = 0
for x in humidity[:10]:
    if day==0:
        day+=1
        print(f"Day:{day}, {x.text}")
    else:
        print(f"Day:{day}, Day:{humidity[itr].text} - Night:{humidity[itr+1].text}, ")
        itr+=1
    itr+=1
    day+=1
    
```
